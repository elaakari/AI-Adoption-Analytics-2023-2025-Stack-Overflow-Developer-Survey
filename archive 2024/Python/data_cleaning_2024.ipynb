```python
columns_map = {
    "Language": [
        "LanguageHaveWorkedWith",
        "LanguageWantToWorkWith",
        "LanguageAdmired"
    ],
    "Database": [
        "DatabaseHaveWorkedWith",
        "DatabaseWantToWorkWith",
        "DatabaseAdmired"
    ],
    "Platform": [
        "PlatformHaveWorkedWith",
        "PlatformWantToWorkWith",
        "PlatformAdmired"
    ],
    "Tools": [
        "ToolsTechHaveWorkedWith",
        "ToolsTechWantToWorkWith",
        "ToolsTechAdmired"
    ],
    "CollabTools": [
        "NEWCollabToolsHaveWorkedWith",
        "NEWCollabToolsWantToWorkWith",
        "NEWCollabToolsAdmired"
    ]
}
```

```python
def build_table(df, category, columns):
    frames = []

    for col in columns:
        if "HaveWorked" in col:
            typ = "Used"
        elif "WantToWork" in col:
            typ = "Wanted"
        else:
            typ = "Admired"

        tmp = (
            df[col]
            .dropna()
            .str.split(';')
            .explode()
            .str.strip()
            .value_counts()
            .reset_index()
        )

        tmp.columns = ["Item", "Count"]
        tmp["Category"] = category
        tmp["Type"] = typ

        frames.append(tmp)

    return pd.concat(frames, ignore_index=True)
```

```python
all_tables = []

for category, cols in columns_map.items():
    all_tables.append(build_table(df, category, cols))

final_df = pd.concat(all_tables, ignore_index=True)
```
