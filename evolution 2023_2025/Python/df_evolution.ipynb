```python
import pandas as pd
df_1 = pd.read_csv(r"..\Data_in\AI_Adoption_2023_Clean.csv",low_memory=False)
df_2 = pd.read_csv(r"..\Data_in\AI_Adoption_2024_Clean.csv",low_memory=False)
df_3 = pd.read_csv(r"..\Data_in\AI_Adoption_2025_Clean.csv",low_memory=False)
```

```python
df_1['Year'] = 2023
df_2['Year'] = 2024
df_3['Year'] = 2025 
```

```python
print("===== 2023 =====")
print(df_1.columns.tolist())

print("\n===== 2024 =====")
print(df_2.columns.tolist())

print("\n===== 2025 =====")
print(df_3.columns.tolist()) 
```

```python
df_3.head(30)
```

```python
df_3['Developer_Profile'] = df_3['MainBranch'].map({
    'I am a developer by profession': 'Professional Developer',
    'I am learning to code': 'Learner',
    'I code primarily as a hobby': 'Hobbyist',
    'I used to be a developer by profession, but no longer am': 'Former Developer',
    'I am not primarily a developer, but I write code sometimes as part of my work/studies': 'Occasional Coder',
    'I work with developers or my work supports developers but am not a developer by profession': 'Developer Support'
})
```

```python
df_3['Developer_Profile'].value_counts()
```

```python
print("===== 2023 =====")
print(df_1.columns.tolist())

print("\n===== 2024 =====")
print(df_2.columns.tolist())

print("\n===== 2025 =====")
print(df_3.columns.tolist()) 
```

```python
common_cols = [
    'Age',
    'Country',
    'Developer_Profile',
    'AI_Usage_Group',
    'Sentiment_Group',
    'Trust_Group',
    'Year'
    
]

df_2023_final = df_1[common_cols]
df_2024_final = df_2[common_cols]
df_2025_final = df_3[common_cols]

df_all = pd.concat(
    [df_2023_final,
     df_2024_final,
     df_2025_final],
    ignore_index=True
)
```

```python
df_all.to_csv(
    r"..\Data_out\AI_Adoption_2023_2025.csv",
    index=False
)
```
