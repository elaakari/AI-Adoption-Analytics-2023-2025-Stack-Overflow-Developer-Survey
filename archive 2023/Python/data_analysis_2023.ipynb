```python
import pandas as pd
df = pd.read_csv(r"..\Data_out\AI_Adoption_2023_Clean.csv",low_memory=False)
```

```python
#What is the AI Adoption Rate in 2023?
valid_users = df[
    df['AI_Usage_Group'] != 'No Response'
]

adoption_rate = (
    (valid_users['AI_Usage_Group'] == 'AI User')
    .mean()
    * 100
)

print(f"AI Adoption Rate: {adoption_rate:.2f}%")
```

```python
#How is AI adoption distributed?
(
    df['AI_Usage_Group']
    .value_counts(normalize=True)
    * 100
).round(2)
```

```python
#What is the overall sentiment toward AI?
(
    df['Sentiment_Group']
    .value_counts(normalize=True)
    * 100
).round(2)
```

```python
#Positive Sentiment Rate
valid_sent = df[
    df['Sentiment_Group'] != 'No Response'
]

positive_rate = (
    (valid_sent['Sentiment_Group'] == 'Positive')
    .mean()
    * 100
)

positive_rate
```

```python
#Do developers trust AI?
(
    df['Trust_Group']
    .value_counts(normalize=True)
    * 100
).round(2)
```

```python
#Trust Rate
valid_trust = df[
    df['Trust_Group'] != 'No Response'
]

trust_rate = (
    (valid_trust['Trust_Group'] == 'Trust')
    .mean()
    * 100
)

trust_rate
```

```python
#BQ5 — Which age groups adopt AI the most?


age_ai = pd.crosstab(
    df['Age'],
    df['AI_Usage_Group']
)

age_ai['AI Adoption Rate'] = (
    age_ai['AI User']
    /
    (
        age_ai['AI User']
        + age_ai['Future User']
        + age_ai['Non User']
    )
) * 100

age_ai[['AI Adoption Rate']]\
.sort_values('AI Adoption Rate', ascending=False)
```

```python
#Which developer profiles adopt AI the most?

profile_ai = pd.crosstab(
    df['Developer_Profile'],
    df['AI_Usage_Group']
)

profile_ai['AI Adoption Rate'] = (
    profile_ai['AI User']
    /
    (
        profile_ai['AI User']
        + profile_ai['Future User']
        + profile_ai['Non User']
    )
) * 100

profile_ai[['AI Adoption Rate']]\
.sort_values('AI Adoption Rate', ascending=False)
```

```python
#Does AI adoption influence sentiment?
pd.crosstab(
    df['AI_Usage_Group'],
    df['Sentiment_Group'],
    normalize='index'
).round(3) * 100
```

```python
#Does AI adoption influence trust?
pd.crosstab(
    df['AI_Usage_Group'],
    df['Trust_Group'],
    normalize='index'
).round(3) * 100
```

```python
#Which countries have the highest AI adoption?

top_countries = (
    df['Country']
    .value_counts()
    .head(15)
    .index
)

country_df = df[
    df['Country'].isin(top_countries)
]

country_ai = pd.crosstab(
    country_df['Country'],
    country_df['AI_Usage_Group']
)

country_ai['AI Adoption Rate'] = (
    country_ai['AI User']
    /
    (
        country_ai['AI User']
        + country_ai['Future User']
        + country_ai['Non User']
    )
) * 100

country_ai[['AI Adoption Rate']]\
.sort_values('AI Adoption Rate', ascending=False)
```

```python
#Which age groups are the most positive toward AI?

sentiment_age = pd.crosstab(
    df['Age'],
    df['Sentiment_Group'],
    normalize='index'
).round(3) * 100

sentiment_age[['Positive']]\
.sort_values('Positive', ascending=False)
```

```python
#Which developer profiles trust AI the most?
trust_profile = pd.crosstab(
    df['Developer_Profile'],
    df['Trust_Group'],
    normalize='index'
).round(3) * 100

trust_profile[['Trust']]\
.sort_values('Trust', ascending=False)
```

```python
#KPI Summary Table
kpis = {
    'Total Respondents': len(df),
    'AI Adoption Rate': round(adoption_rate,2),
    'Positive Sentiment Rate': round(positive_rate,2),
    'Trust Rate': round(trust_rate,2)
}

pd.DataFrame(
    kpis.items(),
    columns=['KPI','Value']
)
```
