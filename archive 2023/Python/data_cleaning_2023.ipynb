```python
important_cols =    [
     "Age",
    "Country",
    "MainBranch",
    "AISelect",
    "AISent",
    "AIBen"

]

df_clean = df[important_cols].copy()
df_clean['Country'] = df_clean['Country'].fillna('Unknown Country')

df_clean['AISelect'] = df_clean['AISelect'].fillna('No Response')

df_clean['AISent'] = df_clean['AISent'].fillna('No Response')

df_clean['AIBen'] = df_clean['AIBen'].fillna('No Response')
```

```python
usage_map = {
    'Yes':'AI User',
    'No, but I plan to soon':'Future User',
    "No, and I don't plan to":'Non User',
    'No Response':'No Response'
}

df_clean['AI_Usage_Group'] = (
    df_clean['AISelect']
    .map(usage_map)
)
```

```python
sentiment_map = {
    'Very favorable':'Positive',
    'Favorable':'Positive',

    'Indifferent':'Neutral',
    'Unsure':'Neutral',

    'Unfavorable':'Negative',
    'Very unfavorable':'Negative',

    'No Response':'No Response'
}

df_clean['Sentiment_Group'] = (
    df_clean['AISent']
    .map(sentiment_map)
)
```

```python
trust_map = {
    'Highly trust':'Trust',
    'Somewhat trust':'Trust',

    'Neither trust nor distrust':'Neutral',

    'Somewhat distrust':'Distrust',
    'Highly distrust':'Distrust',

    'No Response':'No Response'
}

df_clean['Trust_Group'] = (
    df_clean['AIBen']
    .map(trust_map)
)
```

```python
mainbranch_map = {
    "I am a developer by profession":
        "Professional Developer",

    "I am not primarily a developer, but I write code sometimes as part of my work/studies":
        "Occasional Developer",

    "I am learning to code":
        "Learning Developer",

    "I code primarily as a hobby":
        "Hobbyist",

    "I used to be a developer by profession, but no longer am":
        "Former Developer",

    "None of these":
        "Other"
}

df_clean["Developer_Profile"] = (
    df_clean["MainBranch"]
    .map(mainbranch_map)
)
```

```python
columns_map = {
    "Language": [
        "LanguageHaveWorkedWith",
        "LanguageWantToWorkWith"
    ],
    "Database": [
        "DatabaseHaveWorkedWith",
        "DatabaseWantToWorkWith"
    ],
    "Platform": [
        "PlatformHaveWorkedWith",
        "PlatformWantToWorkWith"
    ],
    "Tools": [
        "ToolsTechHaveWorkedWith",
        "ToolsTechWantToWorkWith"
    ],
     "AITools": [
        "AISearchHaveWorkedWith",
        "AISearchWantToWorkWith"
    ],

    "CollabTools": [
        "NEWCollabToolsHaveWorkedWith",
        "NEWCollabToolsWantToWorkWith"
    ]
}
```

```python
def build_table(df, category, columns):
    frames = []

    for col in columns:
        if "HaveWorked" in col:
            typ = "Used"
        else:
            typ = "Wanted"
        

        tmp = (
            df[col]
            .dropna()
            .str.split(';')
            .explode()
            .str.strip()
            .value_counts()
            .reset_index()
        )

        tmp.columns = ["Item", "Count"]
        tmp["Category"] = category
        tmp["Type"] = typ

        frames.append(tmp)

    return pd.concat(frames, ignore_index=True)
```

```python
all_tables = []

for category, cols in columns_map.items():
    all_tables.append(build_table(df, category, cols))

final_df = pd.concat(all_tables, ignore_index=True)
```
