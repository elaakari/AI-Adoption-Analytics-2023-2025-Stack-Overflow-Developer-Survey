```python
general_info_col=[
    "Age",
    "Country",
    "Employment",
    "MainBranch",
    "AISelect",
    "AISent",
    "AIAcc",
    "AIComplex",
    "AIAgents"
]
for col in general_info_col:
    if col in df.columns:
        print("\n")
        print("="*50)
        print(df[col].value_counts(dropna=False).head(20))
            
```

```python
important_cols = [
    'Age',
    'Country',
    'Employment',
    'MainBranch',
    'AISelect',
    'AISent',
    'AIAcc',
    'AIComplex',
    'AIAgents'
]

df_clean = df[important_cols].copy()

df_clean['Age'] = df_clean['Age'].fillna('Unknown')
df_clean['Country'] = df_clean['Country'].fillna('Unknown Country')
df_clean['Employment'] = df_clean['Employment'].fillna('Unknown')
df_clean['MainBranch'] = df_clean['MainBranch'].fillna('Unknown')

for col in ['AISelect', 'AISent', 'AIAcc', 'AIComplex', 'AIAgents']:
    df_clean[col] = df_clean[col].fillna('No Response')
```

```python
general_info_col_clean=[
    "Age",
    "Country",
    "Employment",
    "MainBranch",
    "AISelect",
    "AISent",
    "AIAcc",
    "AIComplex",
    "AIAgents"
]
for col in general_info_col:
    if col in df_clean.columns:
        print("\n")
        print("="*50)
        print(df_clean[col].value_counts(dropna=False).head(20))
```

```python
adoption_map = {
    'Yes, I use AI tools daily':'Daily User',
    'Yes, I use AI tools weekly':'Weekly User',
    'Yes, I use AI tools monthly or infrequently':'Occasional User',
    'No, but I plan to soon':'Future User',
    "No, and I don't plan to":'Non User',
    'No Response':'No Response'
}

df_clean['AI_Usage_Group'] = df_clean['AISelect'].map(adoption_map)
```

```python
sentiment_map = {
    'Very favorable':'Positive',
    'Favorable':'Positive',
    'Indifferent':'Neutral',
    'Unsure':'Neutral',
    'Unfavorable':'Negative',
    'Very unfavorable':'Negative',
    'No Response':'No Response'
}

df_clean['Sentiment_Group'] = df_clean['AISent'].map(sentiment_map)
```

```python
trust_map = {
    'Highly trust':'Trust',
    'Somewhat trust':'Trust',
    'Neither trust nor distrust':'Neutral',
    'Somewhat distrust':'Distrust',
    'Highly distrust':'Distrust',
    'No Response':'No Response'
}

df_clean['Trust_Group'] = df_clean['AIAcc'].map(trust_map)
```

```python
complexity_map = {
    'Very well at handling complex tasks':'Good',
    'Good, but not great at handling complex tasks':'Good',
    'Neither good or bad at handling complex tasks':'Neutral',
    'Bad at handling complex tasks':'Poor',
    'Very poor at handling complex tasks':'Poor',
    "I don't use AI tools for complex tasks / I don't know":'Unknown',
    'No Response':'No Response'
}

df_clean['AI_Capability_Group'] = df_clean['AIComplex'].map(complexity_map)
```

```python
agent_map = {
    'Yes, I use AI agents at work daily':'Active User',
    'Yes, I use AI agents at work weekly':'Active User',
    'Yes, I use AI agents at work monthly or infrequently':'Active User',
    'No, but I plan to':'Future User',
    "No, and I don't plan to":'Non User',
    'No, I use AI exclusively in copilot/autocomplete mode':'Copilot Only',
    'No Response':'No Response'
}

df_clean['Agent_Adoption_Group'] = df_clean['AIAgents'].map(agent_map)
```

```python
model_mapping = {
    "openAI GPT (chatbot models)": "OpenAI",
    "openAI Reasoning models": "OpenAI",
    "openAI Image generating models": "OpenAI",
    "OpenAI Reasoning": "OpenAI",

    "Anthropic: Claude Sonnet": "Claude",

    "Gemini (Flash general purpose models)": "Gemini",
    "Gemini (Pro Reasoning models)": "Gemini",

    "DeepSeek (R- Reasoning models)": "DeepSeek",
    "DeepSeek (V- General purpose models)": "DeepSeek",

    "Meta Llama (all models)": "Llama",
    "X Grok models": "Grok",
    "Mistral AI models": "Mistral",
    "Perplexity Sonar models": "Perplexity",
    "Alibaba Cloud Qwen models": "Qwen",
    "Microsoft Phi-4 models": "Phi-4",
    "Amazon Titan models": "Titan",
    "Cohere: Command A": "Command A",
    "Reka (Flash 3 or other Reka models)": "Reka"
}
```

```python
columns = [
    "AIModelsHaveWorkedWith",
    "AIModelsWantToWorkWith",
    "AIModelsAdmired"
]

for col in columns:
    for old, new in model_mapping.items():
        df[col] = df[col].str.replace(old, new, regex=False)
```

```python
def count_models(column):
    counts = (
        df[column]
        .dropna()
        .str.split(';')
        .explode()
        .str.strip()
        .replace(model_mapping)
        .value_counts()
    )
    return counts
```

```python
worked = count_models("AIModelsHaveWorkedWith")
wanted = count_models("AIModelsWantToWorkWith")
admired = count_models("AIModelsAdmired")
print(worked)
print(wanted)
print(admired)
```
