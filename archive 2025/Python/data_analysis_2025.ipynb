```python
import pandas as pd
df = pd.read_csv(r"C:\Users\PC\Desktop\project\Ai Adoption\archive 2025\AI_Adoption_2025_Clean.csv",low_memory=False)
```

```python
df.head(30)
```

```python
#How frequently do developers use AI?
usage_freq = (
    df['AI_Usage_Group']
    .value_counts(normalize=True)*100
)

usage_freq
```

```python
#Adoption rate
valid = df[df['AI_Usage_Group'] != 'No Response']

ai_users = valid[
    valid['AI_Usage_Group'].isin(
        ['Daily User','Weekly User','Occasional User']
    )
]

adoption_rate = len(ai_users) / len(valid) * 100

print(f"AI Adoption Rate: {adoption_rate:.2f}%")
```

```python
#How do developers feel about AI?
sentiment = (
    df['Sentiment_Group']
    .value_counts(normalize=True)*100
)

sentiment
```

```python
#Positive Sentiment Rate
valid_sent = df[df['Sentiment_Group'] != 'No Response']

positive_rate = (
    (valid_sent['Sentiment_Group'] == 'Positive')
    .mean()*100
)

print(f"Positive Sentiment Rate: {positive_rate:.2f}%")
```

```python
#Do developers trust AI?
trust = (
    df['Trust_Group']
    .value_counts(normalize=True)*100
)

trust
```

```python
#Trust Rate
valid_trust = df[df['Trust_Group'] != 'No Response']

trust_rate = (
    (valid_trust['Trust_Group'] == 'Trust')
    .mean()*100
)

trust_rate
print(f"Trust Rate: {trust_rate:.2f}%")
```

```python
#Do AI users trust AI more?
pd.crosstab(
    df['AI_Usage_Group'],
    df['Trust_Group'],
    normalize='index'
).round(3)*100
```

```python
#Are AI agents already being adopted?

agent_usage = (
    df['Agent_Adoption_Group']
    .value_counts(normalize=True)*100
)

agent_usage
```

```python
#Agent Adoption Rate
valid_agents = df[
    df['Agent_Adoption_Group'] != 'No Response'
]

agent_rate = (
    (valid_agents['Agent_Adoption_Group'] == 'Active User')
    .mean()*100
)

agent_rate
print(f"Agent Adoption Rate: {agent_rate:.2f}%")
```

```python
#How capable is AI for complex tasks?
complexity = (
    df['AI_Capability_Group']
    .value_counts(normalize=True)*100
)

complexity
```

```python
#Which age groups adopt AI the most?

pd.crosstab(
    df['Age'],
    df['AI_Usage_Group'],
    normalize='index'
).round(3)*100
```

```python
#How capable is AI for complex tasks?
age_group = (
    df['Age']
    .value_counts(normalize=True)*100
)

age_group
```

```python
#Which age groups adopt AI the most? only users 

ai_users_only = df[
    df['AI_Usage_Group'].isin([
        'Daily User',
        'Weekly User',
        'Occasional User'
    ])
]

pd.crosstab(
    ai_users_only['Age'],
    ai_users_only['AI_Usage_Group'],
    normalize='index'
).round(3) * 100
```

```python
#Adoption rate by age:

adoption_by_age = pd.crosstab(
    df['Age'],
    df['AI_Usage_Group']
)

adoption_by_age['AI Adoption Rate'] = (
    (
        adoption_by_age['Daily User']
        + adoption_by_age['Weekly User']
        + adoption_by_age['Occasional User']
    )
    /
    (
        adoption_by_age['Daily User']
        + adoption_by_age['Weekly User']
        + adoption_by_age['Occasional User']
        + adoption_by_age['Future User']
        + adoption_by_age['Non User']
    )
) * 100

adoption_by_age[['AI Adoption Rate']].sort_values(
    'AI Adoption Rate',
    ascending=False
)
```

```python
#Does employment status influence AI adoption?
pd.crosstab(
    df['Employment'],
    df['AI_Usage_Group'],
    normalize='index'
).round(3)*100
```

```python
#Are AI users also adopting AI agents?
pd.crosstab(
    df['AI_Usage_Group'],
    df['Agent_Adoption_Group'],
    normalize='index'
).round(3)*100
```

```python
#Which countries have the highest AI adoption?

pd.crosstab(
    df['Country'],
    df['AI_Usage_Group'],
    normalize='index'
).round(3)*100
#Top 20 countries only
top_countries = (
    df['Country']
    .value_counts()
    .head(20)
    .index
)

country_analysis = pd.crosstab(
    df[df['Country'].isin(top_countries)]['Country'],
    df[df['Country'].isin(top_countries)]['AI_Usage_Group'],
    normalize='index'
)

country_analysis.round(3)*100
```

```python
#KPI Summary Table
kpis = {
    'Total Respondents': len(df),
    'AI Adoption Rate': round(adoption_rate,2),
    'Positive Sentiment Rate': round(positive_rate,2),
    'Trust Rate': round(trust_rate,2),
    'Agent Adoption Rate': round(agent_rate,2)
}

pd.DataFrame(
    kpis.items(),
    columns=['KPI','Value']
)
```
